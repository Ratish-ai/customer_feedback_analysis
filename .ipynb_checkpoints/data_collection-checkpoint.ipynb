```python
ratings = []
reviews = []
```

```python
data=html.find_all("div",class_="col _2wzgFH")
```

```python
for i in data:
    rating = i.find_all("div",class_="_3LWZlK _1BLPMq")
    for j in rating:
        ratings.append(j)
    review = i.find_all("div",class_="")
    for j in review:
        reviews.append(j)
```
